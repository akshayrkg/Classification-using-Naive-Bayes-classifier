```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.cross_validation import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
dataset = pd.read_csv("/Users/kaku/Downloads/AIML/git/PIMA/pima-indians-diabetes.csv", header=None)
#dataset.count()
#dataset.describe()
(dataset.head())
```

```python
data=dataset.values[:,:8]
labels=dataset.values[:,-1]
data.shape,labels.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(data,labels,test_size=0.2,random_state=50)
```

```python
#data[:,0].max()
```

```python
clf=GaussianNB()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
accuracy=accuracy_score(y_test, y_pred, normalize=True)
accuracy
```
